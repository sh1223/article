```python
from nltk.corpus import PlaintextCorpusReader
import nltk.corpus
import pandas as pd
import matplotlib.pyplot as plt
```

```python
corpus_root = 'C:/Users/Seonghwan/Desktop/nltk'
textList = PlaintextCorpusReader(corpus_root, '.*')
textList.fileids()
```

```python
len(set(textList.words('macron_gilet_jaune.txt')))
```

```python
fdist=nltk.FreqDist(i.lower() for i in textList.words('macron_gilet_jaune.txt'))
print(fdist)
print(fdist.most_common(710), end='')
```

```python
for i in fdist.most_common(710):
    print(i)
```

```python
listMotSupprimer=" ’ , . de et la le je que nous des pour les l est qui d à une notre en plus c ne pas veux ; n un ce il vous se dans qu par au a tous y cette - mais aux sans du sur : nos ces on s même aussi ont parce celle être dès avons était peut ses j tout cela juste leur où devant avec doit sommes mieux trop soit ceux été devons ils aucune dont seul autres peuvent ait doivent rien aura beaucoup là elle loin fois sont depuis part m mon son ni pourquoi devoir ou chacun donc tant entre abord toutes ai comme toute souvent très me sais ma aucun ainsi sa votre cet moins ? soient moi voulons".split()
listMotSupprimer
```

```python
fdist_net=[i for i in fdist.most_common(710) if i[0] not in listMotSupprimer]
fdist_net
```

```python
mot_plus_deux=[x for x in fdist_net if x[1]>=2]
print(mot_plus_deux)
print('-'*40)
mot_plus_trois=[x for x in fdist_net if x[1]>=3]
print(mot_plus_trois)
```

```python
axe_y=[x[0] for x in mot_plus_trois]
axe_x=[x[1] for x in mot_plus_trois]
print(axe_y)
df_fdist=pd.DataFrame(axe_x, index=axe_y, columns=['fdist'])
df_fdist.plot.bar()
```

```python
df_fdist=pd.DataFrame(axe_x, index=axe_y, columns=['Fréquence'])
df_fdist.plot(kind='barh', fontsize=12, figsize=(7,7))
```
